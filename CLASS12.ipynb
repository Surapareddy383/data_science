```python
import matplotlib.pyplot as plt
print("B\n",np.linspace(2.0,3.0,num=5),"\n")
x=np.linspace(0,2,10)
print("A\n",np.sin(x))
def sinplot(flip=1):
    x=np.linspace(0,14,100)
    for i in range(1,10):
        plt.plot(x,np.sin(x+i*.5)*(7-i)*flip)
sinplot()
plt.show()
```

```python
np.random.seed(10)
sns.set(color_codes=False)
a,c=[0,2],[(1,.5),(.5,1)]
x,y=np.random.multivariate_normal(a,c,size=50).T
ax=sns.kdeplot(x,shade=True,color='blue')
ax=sns.kdeplot(y,shade=True,color='green')
```

```python
ay=sns.kdeplot(x,y,shade=True,color='green')
```

```python
iris=sns.load_dataset('iris')
ax=iris.loc[iris.species=='setosa']
ay=iris.loc[iris.species=='virginica']
x=sns.kdeplot(ax.sepal_width,ax.sepal_length,shade_lowest=False,color='blue',shade=True)
y=sns.kdeplot(ay.sepal_width,ay.sepal_length,shade_lowest=False,color='green',shade=True)
```
