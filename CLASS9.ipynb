```python
x=[3,3,3,3,3]
y=[6,7,8,9,10]
x1=[6,6,6,6,6]
y1=[6,7,8,9,10]
plt.plot(x,y,ls='dashed',color='r',label='first element graph')
plt.plot(x1,y1,color='g',label='second element garph',ls='dashed')
plt.legend(loc='upper right')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[6,7,8,9,10]
z=[2,4,7,9,4]
x1=[11,12,13,14,15]
y1=[15,14,13,12,11]
z1=[1,3,4,4,5]
plt.plot(x,y,ls='dashed',color='r',label='first element graph')
plt.plot(x1,y1,color='g',label='second element garph')
plt.plot(z,z1,color='b',label='third element graph')
plt.legend(loc='lower right')
plt.show()
```
